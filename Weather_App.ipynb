```python
                                        #  Weather_App
import tkinter as tk 
from tkinter import ttk
from tkinter import messagebox,simpledialog,scrolledtext
import requests
import pycountry


class Weather_App():
    def __init__(self):
        self.T=tk.Tk()
        self.T.title("Weather_App")
        self.frame=ttk.Frame(self.T)
        self.frame.pack()
        label_1=tk.Label(self.frame,text="Enter the zip_code/postal_code",font=("Times New Roman", 12))
        label_1.grid(row=2,column=0,padx=10,pady=10)
        zip_postal_entry=tk.Entry(self.frame,width=30)
        zip_postal_entry.grid(row=2,column=1,padx=10,pady=10)
        label_2=tk.Label(self.frame,text="Enter the county name",font=("Times New Roman", 12))
        label_2.grid(row=0,column=0,padx=10,pady=10)
        county=tk.Entry(self.frame,width=30)
        county.grid(row=0,column=1,padx=10,pady=10)
        button=tk.Button(self.frame,text="Get_Weather_info",font=("Times New Roman", 12),command=lambda:self.get_weather(zip_postal_entry,county))
        button.grid(row=4,columnspan=4,padx=10,pady=10)
        self.T.mainloop()
    def cover_temperature(Self,kelvin):
        celsius = kelvin - 273.15
        fahrenheit = (celsius * 9/5) + 32
        return celsius, fahrenheit
    def get_weather(self,zip_postal_entry,county):
        if not zip_postal_entry.get() and county.get() :
            message="Field cant be empty"
            simpledialog.messagebox.showinfo("Error",message)
            return
        if not county.get().isalpha():
            message="county name should only be letters"
            simpledialog.messagebox.showinfo("Error",message)
            county.delete(0,tk.END)
            return
        if not zip_postal_entry.get().isdigit():
            message="Zipcode should be only numbers"
            simpledialog.messagebox.showinfo("Error",message)
            zip_postal_entry.delete(0,tk.END)
            return
        zip_code=zip_postal_entry.get()
        API_key="28a8dc79d04ac9141c46ec2a08b60122"
        country_name=county.get()
        try:
            country_code = pycountry.countries.lookup(country_name).alpha_2
        except LookupError:
            message = "Invalid country name"
            simpledialog.messagebox.showinfo("Error", message)
            return
        url=f"https://api.openweathermap.org/data/2.5/weather?zip={zip_code},{country_code}&appid={API_key}"
        response=requests.get(url).json()
        temperature=response['main']['temp']
        celsius, fahrenheit=self.cover_temperature(temperature)
        humidity=response['main']['humidity']
        temperature_minimum=response['main']['temp_min']
        celsius_min, fahrenheit_min=self.cover_temperature(temperature_minimum)
        temperature_max=response['main']['temp_max']
        celsius_max, fahrenheit_max=self.cover_temperature(temperature_max)
        weather_condition=response['weather'][0]['description']
        location=response['name']
        output_text=f"At the current location is {location}\n The temperature is {celsius:.2f}°C or {fahrenheit}°f\n The maximum temperature is {celsius_max:.2f}°C or {fahrenheit_max}°f\n The minimum_temperature is {celsius_min:.2f}°C or {fahrenheit_min}\n  Humidity={humidity}\n  weathercondition is like{weather_condition}"
        output=scrolledtext.ScrolledText(self.frame,wrap=tk.WORD,width=60, height=20,font=("Arial", 12))
        output.grid(row=6,columnspan=4,padx=10,pady=10)
        output.insert(tk.END, output_text)
        output.config(state="disabled")      
if __name__ == "__main__":
    Weather_App()
```
